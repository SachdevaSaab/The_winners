```python
# Import Dependencies

import pandas as pd
import numpy as np
import requests
import json

# Google API Key
from config import key_list
```

```python
data_csv = ("data/2015.csv")
data1 = pd.read_csv(data_csv)
```

```python
data1.head()
```

```python
# clean data 

data_151 = data1.rename(columns={'Happiness Rank': 'Happiness_Rank', 'Happiness Score': 'Happiness_Score', 'Standard Error': 'Standard_Error','Health (Life Expectancy)': 'Health_(Life_Expectancy)','Standard Error': 'Standard_Error', 'Economy (GDP per Capita)' : 'Economy_(GDP_per_capita)', 'Trust (Government Corruption)': 'Trust_(Goverment_Corruption)', 'Dystopia Residual': 'Dystopia_Residual'})
data_15 = data_151.sort_values('Country', ascending=True)
data_15.reset_index(inplace=True)
data_15.drop(columns={'index'}, inplace=True)

# Add columns for lat, lng

data_15["Lat"] = ""
data_15["Lng"] = ""
data_15["Happiness_Year"] = 2015
data_15.head()
```

```python
g = 0
params = {"key": key_list[0]}
i = 0
# Loop through the cities_pd and run a lat/long search for each city
for index, row in data_15.iterrows():
    try:
        #i += 1
        base_url = "https://maps.googleapis.com/maps/api/geocode/json"

        country = row['Country']
        
        # update address key value
        params['address'] = f"{country}"

        # make request, print url
        data15_lat_lng = requests.get(base_url, params=params)
        print(data15_lat_lng.url)

        # convert to json
        data15_lat_lng = data15_lat_lng.json()

        data_15.set_value(
            index, "Lat", data15_lat_lng["results"][0]["geometry"]["location"]["lat"])
        data_15.set_value(
            index, "Lng", data15_lat_lng["results"][0]["geometry"]["location"]["lng"])
        #print(i)
    except IndexError:
        g += 1
        params = {"key": key_list[g]}
        continue
        
```

```python
data_15.to_csv("data_15.csv", sep=",", index=None)
data_15.to_json('data_15.json')
```
